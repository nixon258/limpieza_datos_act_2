```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/online_retail_dirty_maestria.csv')
```

antes que nada voy a buscar si hay filas duplicadas para eliminarlas

```python
duplicates = df[df.duplicated(keep=False)]
display(duplicates.sort_values(by=['InvoiceNo', 'StockCode']).head(10))
```

como el resultado es que si, entonces vamos a dejar 1 de cada una

```python
df.drop_duplicates(inplace=True)
```

```python
df.head(5)
```

con el df.head pudimos ver las primeras 5 filas y encontramos que la columna quantity tiene datos sin asignar desde el principio, entonces vamos a hacer un df.info para ver mejor todo

```python
df.info()
```

con el df.info encontramos mucho mas, lo mas relevante es que vimos que la columna customerid tiene muchos datos nulos entonces la vamos a eliminar porque igual es una columna que no consideramos muy relevante

```python
df = df.drop('CustomerID', axis=1)
df.head()
```

ahora vamos a limpiar la columna de unitprice, ya que esta como object y no como int o float, para eso buscamos los valores que no son numericos

```python

df[pd.to_numeric(df['UnitPrice'], errors='coerce').isna()]['UnitPrice'].unique()
```

es porq tienen el simbolo "£", entonces lo voy a eliminar.

```python
df['UnitPrice'] = df['UnitPrice'].astype(str).str.replace('£', '', regex=False)
```

y voy a convertir esa columna en float y luego voy a verificar si hay valores en 0

```python
df['UnitPrice'] = pd.to_numeric(df['UnitPrice'], errors='coerce')
```

```python
zero_prices = df[df['UnitPrice'] == 0]
print(f"Cantidad de registros con UnitPrice igual a 0: {len(zero_prices)}")
display(zero_prices.head())
```

note que hay nulos en otras filas entonces voy a mirar cuantos hay y en que filas, asi puedo ver mejor y decido si elimino las filas que unitprice este en 0

```python

zero_prices = df[df['UnitPrice'] == 0]
print(f"Análisis de nulos en registros con UnitPrice 0:")
zero_prices.isnull().sum()
```

```python
df = df[df['UnitPrice'] > 0]
```

Ahora enfoquemonos en la columan quantity que es otra que tiene muchos valores nulos tambien

```python
null_count = df['Quantity'].isnull().sum()
display(df[df['Quantity'].isnull()].head())
```

voy a ver cuales valores son negativos y los convierto en positivos ya que probablemente fue un error de digitacion

```python

negative_quantities = df[df['Quantity'] < 0]
print(f"Cantidad de registros con valores negativos: {len(negative_quantities)}")
```

```python
df['Quantity'] = df['Quantity'].abs()
```

aqui miramos los nulos que coincidian con las variables unitprice y description a ver si tenian correlacion fuerte, pero nos dimos cuenta que no, entonces decidimos borrar las filas de nulos en quantity para no introducir ningun sesgo a los datos en caso de aplicar la media

```python
df[df['Quantity'].isnull()][['UnitPrice', 'Description']].isnull().sum()
```

```python
df = df.dropna(subset=['Quantity'])
df['Quantity'].isnull().sum()
```

tambien voy a cambiar el tipo de dato de la columna invoicedate

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')
```

ahora en cuanto a la columna country encontramos que hay datos repetidos pero escritos de distinta manera (los de United Kingdom), entonces vvamos a convertir todos los que corresponden a united kingdom en uno solo

```python
df['Country'].unique()
```

```python
uk_variants = ['United  Kingdom', 'U.K.', 'UK', 'united kingdom']
df['Country'] = df['Country'].replace(uk_variants, 'United Kingdom')
df['Country'].unique()
```

```python
print(f"Number of 'Unspecified' values in 'Country' column: {df[df['Country'] == 'Unspecified'].shape[0]}")
```

```python
import plotly.express as px

# Agrupamos por país para el mapa
country_map_data = df.groupby('Country')['Quantity'].sum().reset_index()

# Creamos el mapa interactivo
fig = px.choropleth(country_map_data,
                    locations='Country',
                    locationmode='country names',
                    color='Quantity',
                    hover_name='Country',
                    title='Ventas Globales por País (Mapa Interactivo)',
                    color_continuous_scale=px.colors.sequential.Plasma)

fig.show()
```
